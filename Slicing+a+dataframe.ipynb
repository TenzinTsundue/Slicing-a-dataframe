# Slicing a Dataframe (loc and iloc)

```python
# loc : label based accessing function
# iloc : Position based accessing function
```

```python
import pandas as pd

# Create some variables
trials = range(1, 21)
n_rows = len(trials)
subj_id = [1]*n_rows
group = ['Control']*n_rows
condition = ['Affect']*int((n_rows/2)) + ['Neutral']*int((n_rows/2))

# Create a dictionairy
data = {'Condition':condition, 'Subject_ID':subj_id, 
        'Trial':trials, 'Group':group}

# Create the dataframe
df = pd.DataFrame(data)
df.head(10)
```

```python
df[4:10:1]
```

```python
df.loc[4:9, :]
```

```python
df.loc[4, 'Condition']
```

```python
df.loc[4: 9,['Condition', 'Group']]
```

```python
df.loc[4:10, : 'Subject_ID' ]
```

```python
df.loc[4:10, :]
```

```python
df.iloc[4:10, :]
```

```python
df.iloc[4:11, [0,2,3]]
```
